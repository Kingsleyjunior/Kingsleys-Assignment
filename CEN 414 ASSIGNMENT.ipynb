# ANNUAL ROOTS AND TUBERS TOTAL

```python
import pandas as pd
import numpy as np
import statistics as sta
import matplotlib.pyplot as plt

data = pd.read_csv('main.csv')

for i in range (2011, 2020, 1):
    
    x = data['Country or Area']
    y = data[f'{i-1}']
    
    s_dev = y.std()
    
    maximum_v = y.max()
    minimun_v = y.min()
    Range_v = maximum_v - minimun_v
    
    variance_v = sta.variance(y)

    plt.figure(figsize = (15, 20))
    plt.title(f'RODUCTION OF ROOTS AND TUBERS IN {i-1}',fontsize = 20)
    plt.barh(x, y)
    plt.show()
    print(f'The Standard deviation of the bar chart is {s_dev}')
    print(f'The Range is {Range_v}')
    print(f'The Variance is {variance_v}')
```
