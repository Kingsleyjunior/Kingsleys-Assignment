```python
#connect to db
con = psycopg2.connect(
        host = "localhost",
        database="postgres",
        user="postgres",
        password="allow",
        port=5432
)

#cursor
cur=con.cursor()
#CREATE TABLE
create_table = """                  
        create table mytest.population(
              Country varchar NOT NULL, 
              year float NOT NULL, 
              year1 float NOT NULL, 
              year2 float NOT NULL, 
              year3 float NOT NULL, 
              year4 float NOT NULL, 
              year5 float NOT NULL, 
              year6 float NOT NULL, 
              year7 float NOT NULL, 
              year8 float NOT NULL            
         )
     """ #execute query
cur.execute(create_table)

#add values to table (PUT)
def input_values(data):
        try:
                for (index, row) in data.iterrows():
                        insert_sql_query = f'''
                                insert into mytest.population (Country, year, year1, year2, year3, year4, year5, year6, year7, year8)
                                values('{row['Country'].replace("'", " ")}'
                                , '{row['2010']}'
                                , '{row['2011']}'
                                , '{row['2012']}'
                                , '{row['2013']}'
                                , '{row['2014']}'
                                , '{row['2015']}'
                                , '{row['2016']}'
                                , '{row['2017']}'
                                , '{row['2018']}'
                                )'''

                        cur.execute(insert_sql_query)
        except psycopg2.errors.UniqueViolation:
                print('data already exist, add new data')
                pass
data_csv = pandas.read_csv('main.csv')
input_values(data_csv)

#FETCH
cur.execute("SELECT * from mytest.Population")
rows=cur.fetchall()
for r in rows:
        print(r)

#MODIFY
cur.execute("UPDATE mytest.population SET year=100586 WHERE Country='Chad'")
print('Modified items are moved to end of table')

#FETCH
cur.execute("SELECT * from mytest.Population")
rows=cur.fetchall()
for r in rows:
        print(r)
#DELETE TABLE
cur.execute("DROP TABLE mytest.population")
#to apply change
con.commit()
#close crusor
cur.close()
#close connection
con.close()
```
